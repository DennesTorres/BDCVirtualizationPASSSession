```sql
WITH daily_sales AS
(
    select orderdate,count(*) as total
    from FactInternetSales
    where year(orderdate)=2013
    group by orderdate
),
public_holidays AS
(
    SELECT
        holidayname as holiday,
        date
    FROM
        ext.holidays
    WHERE countryorregion = 'United States' AND YEAR(date) = 2013
)
SELECT
[orderdate],
case
  when holiday is null then 0
  else total
end holiday,total
FROM daily_sales s
LEFT OUTER JOIN public_holidays p on s.orderdate = p.date
ORDER BY orderdate ASC;
```
